#### RESEARCH QUESTION 8: How do Listing Prices vary with respect to Bedroom types selected?

```python
rental_totals = df.groupby("original_city")["rentalZestimate"].mean().sort_values()
#rental_totals
rental_totals.plot(kind="barh")
plt.xlabel("Average Price in thousand Dollars - USD")
plt.ylabel("Original Cities")
plt.suptitle("Rental-Price of Cities")
plt.show()
```

is very similar to the Bay area cities of Mountain View, Sunny Vale, Fremont and San Jose

```python
#Code To get Mean Listing Price and Rental with respect to Number of Bedrooms
bed_types = {'0_bed' : ["Studio"],
            '1_bed' : ["1", "1.0"],
            '2_bed' : ["2", "2.0"],
            '3_bed' : ["3", "3.0"],
            '4_bed' : ["4", "4.0"],
            '5_bed' : ["5", "5.0"],
            '5_plus_beds' : ["6", "6.0", "7", "7.0", "8", "8.0"]
            }

bed_wise_mean_price = {} #declared variables for mean listing_price
bed_wise_mean_rental = {} #declared variables for mean rental_price
for (bed_type, value) in bed_types.items():
    bed_wise_mean_price[bed_type] = df[df.beds.isin(value)].groupby("original_city")["price"].mean()
    bed_wise_mean_rental[bed_type] = df[df.beds.isin(value)].groupby("original_city")["rentalZestimate"].mean()
```

#### RESEARCH QUESTION 9: How do listing prices vary with respect to Bedroom types selected?

```python
# Code to get cities and their corresponding mean prices. 
cities = df['original_city'].unique()
for bed_type, citi_wise_mean_price in bed_wise_mean_price.items():
    for city in df['original_city'].unique():
        if city not in citi_wise_mean_price:
            citi_wise_mean_price[city] = 0

#Create datafram from dict
BedType_wise_prices = pd.DataFrame.from_dict(bed_wise_mean_price, orient='index',columns=list(cities))

#Ploting  Listing Price of Cities
plt.figure(figsize=(20,8))

sns.lineplot(
    data=BedType_wise_prices/1e6, dashes=False)
plt.xlabel("Number of bed rooms")
plt.ylabel("price in Million Dollars USD")
plt.suptitle("Listing Price of Cities")
plt.show()
```

Observation: 
For every city the average listing price increases as the number of bedrooms in the property also 
increase (exception of New york)
The cities of New york, LA, San Francisco &  Palo Alto are the top 4 in listing prices


#### RESEARCH QUESTION 10: How do Rental prices vary with respect to Bedroom types selected ?

```python
# Code to get cities and their corresponding mean prices. 
cities = df['original_city'].unique() #for all citites in the dataframe
for bed_type, citi_wise_mean_rent in bed_wise_mean_rental.items():
    for city in df['original_city'].unique():
        if city not in citi_wise_mean_rent:
            citi_wise_mean_rent[city] = 0

#Create datafram from dict
BedType_wise_rental = pd.DataFrame.from_dict(bed_wise_mean_rental, orient='index',columns=list(cities))

#Ploting  Listing Price of Cities
plt.figure(figsize=(20,8))
# plt.yscale('log')
sns.lineplot(
    data=BedType_wise_rental/1e3, dashes=False)
plt.xlabel("Number of bed rooms")
plt.ylabel("price in 1000 Dollars USD")
plt.suptitle("Rental Price of Cities")
plt.show()
```

Observation :
The cities of New york, LA, San Francisco &  Palo Alto are the top 4 in rental price upto 4 bedrooms.

Similar to the listing prices, the rental prices also do not vary significantly for properties with upto 4 bed rooms

# Regression

### Linear Regression

```python
regression_df=df.loc[:,['price','sqft','Price/sqft','ele_rating','mid_rating', 'high_rating','views']]
```

```python
#X are the independent variables, y is the dependent variable
from sklearn.model_selection import train_test_split

X=regression_df.loc[:,['sqft','Price/sqft','ele_rating','mid_rating','high_rating','views']]

y=regression_df['price']

# Using a 70-30 split for training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=5)

print("Training: ", X_train.shape)
print("Test :", X_test.shape)

```

```python
from sklearn.linear_model import LinearRegression

# Linear Regression and model fitting
lr_model1 = LinearRegression()
lr_model1.fit(X_train,y_train)
```

```python
#plot scatterplot of Actual Selling Price vs the Predicted Price by the model

test_set_pred1 = lr_model1.predict(X_test) # Using the linear regression model to predict the price from the test data
plt.figure(figsize=(8, 5))
ax = sns.regplot(x = test_set_pred1, y = y_test) # plotting the predicted price with actual price
plt.title('Actual Selling Price vs Predicted Price ,S')
plt.xlabel('Actual Selling Price,$')
plt.ylabel('Predicted Selling Price,$')
plt.xlim(0, 60000000)

```

```python
#Comparison between Predicted and Actual Prices

result = pd.DataFrame({'Predicted': test_set_pred1, "Actual" : y_test})
print(result.head(20))
```

```python
#Statistics of this model

from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

print("Statistics of Linear Regression")
print("\n")
#RMSE -Root Mean Square Error
print("RMSE: ",np.sqrt(np.mean((test_set_pred1 - y_test)**2)))

# MAD -Mean Absolute Deviation
print("MAD: ",np.mean(np.abs(test_set_pred1 - y_test)))

#MSE -Mean Squared Error
print("MSE: ",mean_squared_error(y_test, test_set_pred1))

print("R2 score: ",r2_score(y_test, test_set_pred1))
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

selected_columns_2 = ['sqft','Price/sqft', 'ele_rating','mid_rating','high_rating','views']

rfmodel1=RandomForestRegressor(n_estimators = 600, min_samples_leaf=1,  n_jobs=-1, max_features='sqrt') #Your parameters inside the parentheses
rfmodel1.fit(X_train.loc[:,selected_columns_2],y_train)
rf1_test_pred = rfmodel1.predict(X_test.loc[:,selected_columns_2])
```

```python
# plotting the predicted price with actual selling price

ax = sns.regplot(y_test, rf1_test_pred) 
plt.title('Actual Selling Price vs Predicted Price, $')
plt.xlabel('Actual Sold-price,$')
plt.ylabel('Predicted Selling Price,$')

```

```python
#Statistics of this model

from sklearn.metrics import r2_score

print("Statistics of Random Forest Regressor")
print('\n')
#RMSE - Root Mean Squared Error
print("RMSE: ",np.sqrt(np.mean((rf1_test_pred - y_test)**2)))

#MAD -Mean Absolute Deviation
print("MAD: ",np.mean(np.abs(rf1_test_pred - y_test)))

#MSE -Mean Squared Error
print("MSE: ",mean_squared_error(y_test, rf1_test_pred))

print("R2 score: ",r2_score(y_test, rf1_test_pred))
```

```python
#Comparison between Predicted and Actual Prices

result = pd.DataFrame({'Predicted': rf1_test_pred, "Actual" : y_test})
print(result.head(20))
```

```python
#Important features which influence the Price

importance = pd.DataFrame({'features': X_train.loc[:,selected_columns_2].columns, 
                           'importance': rfmodel1.feature_importances_})
importance.index=importance.features
importance.sort_values(by='importance', ascending=True).plot.barh()

```

#### Observations:

1. From both models we can infer that the R2 score and the RMSE for the Random Forest Regressor is lesser when compared to Linear Regression.
2. The RMSE gives us more accurate results and so it can be considered as a better model to predict selling prices.
3. The most important features that influenced the prediction of prices is the Price/sqft and price as shown in the plot.
